```python
#option 1: Addition Operator (+)

A = [1,2,3,4,5,6]
B = [13, 21, 34]
A_B = A + B
A_B
```

```python
#option 2: Appending list B into A

A = [1,2,3,4,5,6]
B = [13, 21, 34]
for x in B:
  A.append(x)
print(A)
```

```python
#option 3: Using extend() method to add list B to list A:

A = [1,2,3,4,5,6]
B = [13, 21, 34]
A.extend(B)
print(A)
```
